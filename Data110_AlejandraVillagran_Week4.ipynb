```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")
```

```python
#Task 1

x = ["Aggravated Assault","Burglary","Larceny","Motor Vechile Theft","Murder","Rape","Robbery"] #categories
y = [264.1,250.7,1347.20,318.70,5.70,38.00,65.50] #Values




colors =['orange','blue','red','green','lightblue','yellow','black']
plt.bar(x,y,facecolor = colors)

plt.title('Crime Rates in 2023')
plt.xlabel('Type of Offense')
plt.ylabel('Offenses per 100,000 People')
plt.grid(True)
plt.show()
```

```python
#Task 2

x = ["Aggravated Assault","Burglary","Larceny","Motor Vechile Theft","Murder","Rape","Robbery"] #categories
y = [-3.3,-8.1,-4.9,12.0,-12.0,-9.8,-.8] #Values




colors =['orange','blue','red','green','lightblue','yellow','black']
plt.barh(x,y,facecolor = colors)

plt.title('Change in Crime Rates(2022vs.2023)')
plt.xlabel('Changes in Offenses per 100,000 People')
plt.ylabel('Type of Offense')
plt.grid(True)
plt.show()
```

Why does this bar graph make more sense than Task 1's bar graph?
This bar graph is more insightful than the first one because it provides information about change over time. The first bar graph only showed whole numbers, meaning that the information only illustrated how many offenses were committed that year. However, by including the percentage that shows the crime rate from one year to another, the graph gives more information about trends in crimes.   

```python
#different Version to improve clarity

## Used this resources to order data: https://www.geeksforgeeks.org/seaborn-sort-bars-in-barplot/

import seaborn as sns
x = ["Aggravated Assault","Burglary","Larceny","Motor Vechile Theft","Murder","Rape","Robbery"] #categories
y = [-3.3,-8.1,-4.9,12.0,-12.0,-9.8,-.8] #Values


df = pd.DataFrame({"x":x,"y":y})

df_sorted =df.sort_values('y',ascending=False)

sns.barplot(x='x',y='y', data=df_sorted)


plt.title('Change in Crime Rates(2022vs.2023)')
plt.xlabel('Changes in Offenses per 100,000 People')
plt.xticks(rotation=50)
plt.ylabel('Type of Offense')
plt.grid(True)
plt.show()
```

In this version, I changed the order of the data to improve the readability of the graph. I thought that by changing the order, it would make it easier and clearer to show the drop in the crime rate. I also adjusted the x-axis labels to ensure they have enough space.
