```python
#week 3 assignment
import matplotlib.pyplot as plt;
print(plt.style.available);
plt.style.use('seaborn-v0_8-white');
plt.style.use('darkbackground');

import numpy as np
```

```python
#Basic line Plot
x=np.linspace(-2,4,100)
print(x)

y = x ** 2 - 2 *x +1
plt.plot(x,y)
```

```python
#Multiple Lines
x = np.linspace(-5,10,200)
f = x**3
g = np.exp(x)
h = np.log(x)

plt.xlim(-2,20)
plt.ylim(-5,20)

plt.plot(x,f, label ="f=x^3", color ='k',)
plt.plot(x,g,label="g=e^x", color = 'r')
plt.plot(x,h, label="h=log(x)", color ='b')

plt.legend()
plt.show()
```

```python
#styling Lines
x=np.linspace(0,10,500)
f = x**2
g = x**3
h = x**4
i = x**5

plt.plot(x,f, linestyle ='-',color='g')
plt.plot(x,g, linestyle = '--',color='r')
plt.plot(x,h, linestyle ='-.',color='b')
plt.plot(x,i, linestyle =':',color='k')

plt.xlim(0,9)
plt.ylim(0,200)

plt.show()
```

```python
#Axes Limits and Scaling
x=np.linspace(-10,10,100)

y= 1/(x**2 +1)

plt.plot(x,y)
plt.axis('equal')
plt.show()

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.plot(x,y)
plt.axis('tight')
plt.show()
```

```python
#Labels and Legends
x=np.linspace(1,10,100);

y = x**0.5
g = x**(1/3)

plt.title('f=x^0.5 and g=x^(1/3)')
plt.plot(x,y,label='x^.5')
plt.plot(x,g,label='x^1/3')
plt.legend()
plt.show()
```

```python
#Basic Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.plot(x,y,marker='o')
plt.show()

plt.scatter(x,y,marker='o')
plt.show()
```

```python
#Custom Scatter Plot
x=np.linspace(0,4 *np.pi,100)

y = x * np.sin(x)

plt.scatter(x,y,marker="s",s=15,facecolors='red',edgecolors='green',linewidths=1.5)

plt.plot(x,y,color='k',linewidth=1)
plt.show()
```

```python
#Combined Plot
x=np.linspace(-2 * np.pi,2* np.pi,100)

y = x * np.cos(x)

plt.plot(x,y,color='r')

x_scatter = np.arange(-2 * np.pi,2* np.pi, np.pi / 4)

plt.scatter(x,y, marker='p')
plt.show()
```

```python
#Custom Plot with Alternative Style
plt.style.use('dark_background')

x = np.linspace(-5,10,100)

y = np.tan(x)
i = np.arctan(x)

plt.plot(x,y, linestyle='--',label="y = tanx")
plt.plot(x,i,linestyle=':',label="i = arctanx")


plt.title('Plot of y=tanx and i=arctanx')
plt.legend()
plt.show()

#I chose the functions for y and i because I did not have to worry about their limits like in previouse examples, where log had to be greater than 0.
#The style I choose made it easier for me to see the data points. The dark background allows for more contrast.
```

```python
#Style Comparison
#Using Example from Basic Scatter Plot

#Using default style
plt.style.use('default')
x = np.random.rand(50)
y = np.random.rand(50)

plt.plot(x,y,marker='o')
plt.title('Basic Scatter Plot using Default')
plt.show()

plt.scatter(x,y,marker='o')
plt.title('Basic Scatter Plot using Default')
plt.show()

#using seaborn
plt.style.use('seaborn-v0_8-white')
x = np.random.rand(50)
y = np.random.rand(50)

plt.plot(x,y,marker='o')
plt.title('Basic Scatter Plot using Seaborn White')
plt.show()

plt.scatter(x,y,marker='o')
plt.title('Basic Scatter Plot using Seaborn White')
plt.show()

#using my Chosen Style
plt.style.use('dark_background')
x = np.random.rand(50)
y = np.random.rand(50)

plt.plot(x,y,marker='o')
plt.title('Basic Scatter Plot using Dark Background')
plt.show()

plt.scatter(x,y,marker='o')
plt.title('Basic Scatter Plot using Dark Background')
plt.show()
```

I'm not sure if I used default and seaborn correctly. They both appear to have the same style and look when using the scatter plot — there was no visible change. However, when using the style I chose, 'dark_background', I can immediately see the change in the background color to black. Personally, I feel that the dark background made the data points easier to see. For this specific type of data, I would use the 'dark_background' style as it makes the points more visible.
