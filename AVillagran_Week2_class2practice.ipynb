```python
plt.figure(figsize=(10,6))
plt.scatter(df['Rank'],df['HappinessScore'],s=10,c='red',marker='s')
plt.xlabel('Rank')
plt.ylabel('HappinessScore')
plt.title('Rank vs Happiness')
plt.grid()
plt.show()

#info about scatter plot https://matplotlib.org/stable/api/index.html
```

```python
plt.scatter(df['HappinessScore'],df['Social support'])
plt.xlabel('Happiness Score')
plt.ylabel('Social Support')
plt.title('Happiness vs Social Support')
```
